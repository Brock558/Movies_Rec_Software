```python
# the data is in weired form we have to convert it into a ['Action', 'Adventure', 'Fantasy', Science Fiction']
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
     L.append(i['name'])
    return L
```

```python
# now we have to convert string of list into list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1000)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def cast_converter(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter<3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(cast_converter)
```

```python
movies.head()
```

```python
#movies.iloc[3]['crew']
```

```python
def dir(rtr):
    L = []
    for i in ast.literal_eval(rtr):
        if i['job'] == 'Director':
           L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(dir)
```

```python
movies.head()
```

```python
movies['overview']= movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['crew'] + movies['cast'] + movies['keywords'] + movies['overview'] + movies['genres'] + movies['cast']
```

```python
movies = movies.drop(columns=['crew', 'cast', 'keywords', 'overview', 'genres'])
```

```python
movies['tags']= movies['tags'].apply(lambda x: " ".join(x))
```

```python
cv = CountVectorizer(max_features=5,stop_words='english')
```

```python
vector = cv.fit_transform(movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def Write_movie_name(movie):
    index = movies[movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key= lambda x:x[1])
    for i in distances[1:6]:
        print(movies.iloc[i[0]].title)
```

```python
Write_movie_name('Pirates of the Caribbean: At World's End')
```
